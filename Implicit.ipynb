```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function that represents the differential equation
def dydt(t, y):
    return t + y

# Adams-Bashforth 2-step method implementation
def adams_bashforth_2(f, t0, y0, t1, h):
    # Number of steps
    n = int((t1 - t0) / h) + 1
    
    # Arrays to store t and y values
    t = np.zeros(n)
    y = np.zeros(n)
    
    # Initial condition
    t[0] = t0
    y[0] = y0
    
    # Step 1: Use Euler's method to get the second point
    t[1] = t[0] + h
    y[1] = y[0] + h * f(t[0], y[0])
    
    # Step 2 and beyond: Adams-Bashforth 2-step method
    for i in range(2, n):
        t[i] = t[i-1] + h
        # Adams-Bashforth formula
        y[i] = y[i-1] + (h / 2) * (3 * f(t[i-1], y[i-1]) - f(t[i-2], y[i-2]))
    
    return t, y

# Parameters for integration
t0 = 0
y0 = 1  # Initial condition
t1 = 10  # End time
h = 0.1  # Step size

# Get the results using Adams-Bashforth 2-step method
t, y = adams_bashforth_2(dydt, t0, y0, t1, h)

# Plot the results
plt.plot(t, y, 'b-', label='Adams-Bashforth 2-step')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Adams-Bashforth 2-step Integration')
plt.grid(True)
plt.show()
```
